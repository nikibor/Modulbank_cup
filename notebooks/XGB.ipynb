```python
import pandas as pd
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.base import TransformerMixin
```

```python
data = pd.read_csv('/home/nikita/PycharmProjects/Modulbank/train.csv',sep='	', index_col=0)
test = pd.read_csv('/home/nikita/PycharmProjects/Modulbank/test.csv',sep='	', index_col=0)
data.head()
```

```python
features = data.columns[1:]
big_X = data[features].append(test[features])
# big_X_imputed = DataFrameImputer().fit_transform(big_X)
```

```python
train_X = big_X[0:data.shape[0]].as_matrix()
test_X = big_X[data.shape[0]::].as_matrix()
train_y = data['0']
```

```python
gbm = xgb.XGBClassifier(max_depth=48, n_estimators=345, learning_rate=0.05, booster='dart').fit(train_X, train_y) #0.568
predictions = gbm.predict(test_X)
```

```python
submission = pd.DataFrame({ '_ID_': test.index,
                            '_VAL_': predictions })
submission.to_csv("submission.csv", index=False)
```
